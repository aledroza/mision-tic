# Manipulación de Imagenes
A continuación se presentan algunas funciones básicas para el manejo de imágenes con MatplopLib, El objetivo es explorar el potencial de las funciones de Python, más que profundizar en esta librería en particular, sin embargo en esta  [dirección](https://matplotlib.org/stable/tutorials/index.html) se pueden encontrar todos los tutoriales sobre ella.

Al igual que con el archivo de rostros.txt es necesario tener una imagen guardada en tu servidor collab.

```python
import numpy as np
import matplotlib.cm as cm
import matplotlib.pyplot as plt
import matplotlib.cbook as cbook
from matplotlib.path import Path
from matplotlib.patches import PathPatch

# Una imagen de ejemplo
with cbook.get_sample_data('/content/MIT2021/hacker.jpg') as image_file:
    image = plt.imread(image_file)

fig, ax = plt.subplots()
ax.imshow(image)
ax.axis('off')  # clear x-axis and y-axis
```

# Algunas transformaciones

```python
lum_img = image[:, :, 0]
plt.imshow(lum_img)
```

```python
plt.imshow(lum_img, cmap="hot")
```
